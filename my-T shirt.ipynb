```python
dist={}

for i in range(0,len(x)):
    d=0
    for j in range(0,1):
        a=((x[i][j]-x1[j])**2)
        
    
    for k in range(1,2):    
        b=((x[i][k]-x1[k])**2)
        d=a+b
        
    d=sqrt(d)
    dist[i]=d
    
print(dist)
```

```python
d1=sorted(dist.items(),key=lambda k:k[1])
print(d1)
```

```python
k=15
level=[]
for i in range(0,k):
    z=d1[i][0]
    level.append(y[z])
    
print(level)
```

```python
ll={}

pq=['S','M','L','XL','XXL']

m1=0
m2=0
m3=0
m4=0
m5=0
for i in range (0,len(level)):
    if(level[i]==pq[0]):
        m1=m1+1;
        ll[0]=m1
    elif(level[i]==pq[1]):
        m2=m2+1
        ll[1]=m2
    elif(level[i]==pq[2]):
        m3=m3+1
        ll[2]=m3 
    elif(level[i]==pq[3]):
        m4=m4+1
        ll[3]=m4 
    elif(level[i]==pq[4]):
        m5=m5+1
        ll[4]=m5        
        
#d=max(m1,m2)
print(ll)
```
